# GloVeの単語ベクトルによる類似語検索

```python
from gensim.models import KeyedVectors
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm

glove_vector_file = "gensim_glove_vectors.txt"
word = "カバー"
top_k = 20

fonts = fm.findSystemFonts()
plt.rcParams['font.family'] = 'IPAexGothic'

glove_vectors = KeyedVectors.load_word2vec_format(glove_vector_file, binary=False)
results = glove_vectors.most_similar(word, [], top_k)

rank = np.array([i + 1 for i in range(len(results))])
words = np.array([result[0] for result in results])
score = np.array([result[1] for result in results])

plt.rcParams["font.size"] = 15
plt.figure(figsize=(10,10))
plt.barh(rank, score, color="blue", alpha=0.6)
plt.yticks(rank, words)

plt.title("Top {} similar words for {}".format(top_k, word))
plt.xlabel("Score")
plt.ylabel("Words")
plt.show()

```
